# Instancia para tokenizar el dataset

```python
from transformers import AutoTokenizer
from datasets import load_dataset, load_metric
import numpy as np
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer


model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

actual_task = "boolq"
dataset = load_dataset("super_glue", actual_task)
metric = load_metric('super_glue', actual_task)
sentence1_key, sentence2_key = task_to_keys[actual_task]

def preprocess_function(examples):
    print((examples).keys())
    return tokenizer(examples[sentence1_key], examples[sentence2_key], truncation=True)

def compute_metrics(eval_pred):
    predictions, labels = eval_pred

    predictions = np.argmax(predictions, axis=1).tolist()

    return metric.compute(predictions=predictions, references=labels)

encoded_dataset = dataset.map(preprocess_function, batched=True)
```

```python

batch_size = 16

args = TrainingArguments(
    f"ensamble-finetuned-{actual_task}",
    evaluation_strategy = "epoch",
    do_predict= True,
    learning_rate=1e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=5,
    weight_decay=0.01
)
validation_key = "validation"
trainer = Trainer(
    model_prueba,
    args,
    train_dataset=encoded_dataset["train"],
    eval_dataset=encoded_dataset[validation_key],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train() #ENSAMBLE UWU BOOLQ
# model.push_to_hub("prueba2")
# tokenizer.push_to_hub("prueba2")
```
